```python
"""
Write, Update, and Delete operations in Delta Lake are ACID transactions, and each one creates a new version of the table.

Every change is stored as a new JSON entry in the _delta_log.

Delta writes each version as a separate JSON log file.

Once there are 10 JSON log files, Delta automatically creates a checkpoint (a compacted Parquet file) to speed up reads and metadata processing.

"""
```

```python
# readinf data from volume 

df = spark.read.format("csv") \
    .option("path","/Volumes/avd_workspace1/default/raw_data/employees_1.csv") \
    .option("header","true") \
    .load()
```

```python
df.show()
```

```python
# By default, saveAsTable() creates a managed table, where Databricks manages both the table metadata in the Unity Catalog (or Hive Metastore) and the underlying data files.

df.write.format("delta").saveAsTable("employee")
```

```python
%sql

-- see the current version of table
-- as we have nothing changed in table since we se here a single version 

DESCRIBE HISTORY employee 
```

```python
%sql 

-- update address Pune = Mumbai

update employee set address = "Mumbai" where address = "Pune"
```

```python
%sql

-- see the current version of table
-- as we have update table we can see 2 version one is for create table and 2nd is for update 
-- but we aree seeing 3 becasue it has optimize table by iteself that why we can see 3 version

DESCRIBE HISTORY employee 
```

```python
%sql 

-- deleting from table 

delete from employee where address = "Mumbai"
```

```python
%sql

-- see the current version of table
-- here we can see the new version added for delete operation

DESCRIBE HISTORY employee 
```

```python
%sql
-- we can select spcific version of file useing version as of 
-- here we are selecting version as of 0 meanse - starting version - when we created table 

select * from employee version as of 0
```

```python
%sql
-- we can select spcific version of file useing version as of 
-- here we are selecting version as of 1 meanse - in version 1 we have updated the table

select * from employee version as of 1
```

```python
%sql
-- we can select spcific version of file useing version as of 
-- here we are selecting version as of 4 meanse - in version 3 we have deleted the record but 1 extra version we got for optimiz

select * from employee version as of 4 
```

```python
%sql

-- we can select specific version by using date also - like we have update the record on 2025-12-03T18:44:36.000+00:00 so we can select that version by using date

select * from employee timestamp as of "2025-12-03T18:44:36.000+00:00"
```

```python
%sql 

-- we can also rollback version 
-- in version 2 we have update record 

restore table employee to version as of 2
```
