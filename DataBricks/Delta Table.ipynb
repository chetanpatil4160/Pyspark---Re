```python
%sql 

USE avd_db
```

```python
# saving data in Delta table
# Detla talbe = parquet + detla_log
# in db we can one folder is for log - _delta_log
df.write.format("delta").save("/Volumes/avd_workspace1/avd_db/raw_data/delta_1")
```

```python
# reading data froma that delta table

df2 = spark.read.format("delta").load("/Volumes/avd_workspace1/avd_db/raw_data/delta_1")
df2.show()

# reading form that log folder
d_log = (spark.read.format("json").option("path","/Volumes/avd_workspace1/avd_db/raw_data/delta_1/_delta_log/00000000000000000000.json")
         .load())
```

```python
%sql
-- but every time we can not read that log files for that there is below command

-- we can see below all the history like first we write data , then we update data nd they have optimize it automattically

DESCRIBE HISTORY delta.`/Volumes/avd_workspace1/avd_db/raw_data/delta_1`
```

```python
%sql
--now we delete data froam table 
delete from delta.`/Volumes/avd_workspace1/avd_db/raw_data/delta_1` where id = 3;
```

```python
%sql

-- now we see delete operation there
-- but every time we do any operation it will create 1 json file in log but with operation files goes increasing 
-- that is why it do checkpointing - where file count goes above 10 it will one file with all fils 
DESCRIBE HISTORY delta.`/Volumes/avd_workspace1/avd_db/raw_data/delta_1`
```
